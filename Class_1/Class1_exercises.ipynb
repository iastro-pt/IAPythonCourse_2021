# Complex numbers (Classes)

## Problem description 

Create a class that is able to handle complex numbers and perform some operations. 

Basic functions: 
- Sum, multiply and subtract two complex numbers
- print a (nicely) formatted number (i.e. a + bi)
- Compute the modulus of the number 


## Hint / Help 

When dealing with classes it is possible to overload some basic functions of python, i.e. write our own logic for that operation.

## Useful resources

- Complex numbers: https://en.wikipedia.org/wiki/Complex_number
- overloading operators: https://docs.python.org/3/reference/datamodel.html#emulating-numeric-types


```python
class Value:
    def __init__(self, val):
        self.value = val
    
    def __add__(self, other):
        """
        Sum the two values and return it as a new "Value" object
        """
        return Value(self.value + other.value)

    def __str__(self):
        return "Value class of {}".format(self.value)

foo = Value(1)
bar = Value(2)
foobar = foo + bar 

for item in [foo, bar, foobar]:
    print(item)
```

# Median filter

## Problem description

The need of estimating the continuum level of a signal/data is a problem often faced when dealing with stellar spectra. One possible tool to solve it is a median filter, a technique often used to remove noise from an image or signal. The main idea of this filter filter is to run through the signal entry by entry, replacing each entry with the median of neighboring entries (window centered in each point, size must always be odd).


## Data 

Inside the /data folder we have a fits file containing one observation from HARPS (albeit with a different format from the one that comes from ESO archive). The goal of this exercise is to create a function that implements a median filter with a window of size N, with a given boundary condition, and apply it to each spectral order of the observation. 


## Boundary conditions

To account for boundary conditions **one** of the following (equal to those from scipy) can be selected: 


- reflect (dc b | a b c d | d c b a)

    - The input is extended by reflecting about the edge of the last pixel. This mode is also sometimes referred to as half-sample symmetric.

- nearest (a a a a | a b c d | d d d d)

    - The input is extended by replicating the last pixel.
- mirror (d c b | a b c d | c b a)

    - The input is extended by reflecting about the center of the last pixel. This mode is also sometimes referred to as whole-sample symmetric.
- wrap (a b c d | a b c d | a b c d)

    - The input is extended by wrapping around to the opposite edge.



Then, it will be applied to each spectral order and the output will be compared against the spectra
